```python
# so now lets make 3 variables , Age , Name  ,is_student

Name = input('My name is ')
Age = int(input('my age is '))
is_student_input = input('Are you a student? (yes or no): ')

if is_student_input.lower() == 'yes':
    is_student = 'is a student'
else:
    is_student = 'is not a student'

if Age >= 23:
    print(f'{Name} is 23 or older and {is_student}.')
else:
    print(f'{Name} is younger than 23 and {is_student}.')
```
